```python
import time
def check_pin(pin):
    true_pin = str(time.time_ns()//1000000000//3600//24//7) # nekoukat, tajne!!!
    if len(pin) != len(true_pin):
        return False
    for i in range(len(true_pin)):
        time.sleep(0.02) ######### Tady simulujeme kontrolu aktualni cislice PINu
        if pin[i] != true_pin[i]:
            return False
    return True
%time print(check_pin("1234"))
```

```python
from IPython.display import display,SVG
from ipywidgets import Button, HBox, VBox, Label

pin_string = ""
labels = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '<<', '0', 'OK']
buttons = [Button(description=l) for l in labels]
status = Label(value="Zadej PIN")

def clicked(arg):
    global pin_string, status
    digit = arg.description # ktere tlacitko zmacknuto?
    if digit == "OK":
        start = time.time()
        result = check_pin(pin_string) # kontrola pinu, merime cas
        end = time.time()
        print(f"Pin = {pin_string}: Cas kontroly = {(end - start) * 1000.:6.2f} ms")
        pin_string = ""
        status.value = ["Chybny PIN!", "PIN OK"][result]
        if result:
            display(SVG(filename='secret.bin')) # odmena za uspech
    elif digit == "<<":
        pin_string = pin_string[:-1]
        status.value = pin_string
    else: # skutecna cislice
        pin_string += digit
        status.value = pin_string

for b in buttons:
    b.on_click(clicked)

# Tlacitka budou po 3 v radku, pod tim celym pak stavovy radek
grid = VBox([HBox(buttons[i:i + 3]) for i in range(0, len(labels), 3)] + [status])
# Zobrazime tlacitka a stavovy radek
display(grid)
```
